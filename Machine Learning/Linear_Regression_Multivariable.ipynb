Reading CSV File

```python
df = pd.read_csv('/content/drive/MyDrive/hiring.csv')
df
```

This candidates data has one missing value in test_score column.
Fill the missing value by finding the median of each value in this column.

```python
import math
median_testscore = math.floor(df['test_score(out of 10)'].median())
median_testscore
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_testscore)
df
```

Now we need to convert 'experience' column from strings to float.

- First fill empty values with zeros.

```python
df.experience = df.experience.fillna("zero")
df
```

Assign strings to its corrosponding values



```python
def word_to_num(word):
    word_map = {
        'zero': 0,
        'two': 2,
        'three': 3,
        'five': 5,
        'seven': 7,
        'ten': 10,
        'eleven': 11,
    }
    return word_map.get(word, word)
```

Now, convert strings to float.

```python
df.experience = df.experience.apply(word_to_num)
df
```

Seperating the variables

```python
model_lr = linear_model.LinearRegression()
```

```python
model_lr.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])
```

###Make Pridictions

> 2 year experience, 9 test scores and 6 interviews

```python
model_lr.predict([[2, 9, 6]])
```

- The above is giving predictions of 53205 salary based on the givenn input.
-  However, when we increase the experince and an interview then the salary would be increased as in the next predictions below

> 12 year experience, 10 test scores and 10 interviews

```python
model_lr.predict([[12, 10, 10]])
```
