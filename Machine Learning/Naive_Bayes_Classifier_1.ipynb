#Naive Bayes Classifier

```python
import seaborn as sns #!
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

df_titanic = sns.load_dataset(name='titanic')
df_titanic.head()
```

```python
df_titanic = sns.load_dataset(name='titanic')[['pclass', 'sex', 'age', 'fare', 'survived']]
df_titanic.head()
```

```python
df_titanic.describe()
```

```python
df_titanic.info()
```

```python
target = df_titanic['survived']
inputs = df_titanic.drop(columns='survived')
```

```python
df_titanic = pd.get_dummies(inputs.sex)
df_titanic.head()
```

```python
inputs = pd.concat([inputs, df_titanic], axis='columns')
inputs.head()
```

```python
inputs.drop('sex', axis='columns', inplace=True)
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs['age']
```

```python
inputs.age = inputs.age.fillna(inputs.age.mean())
inputs
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
model_nb = GaussianNB()
```

```python
model_nb.fit(X_train, y_train)
```

```python
model_nb.score(X_test, y_test)
```
