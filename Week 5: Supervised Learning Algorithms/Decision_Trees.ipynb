#Decision Trees

- In the decision tree, if all the occurrances is true or green, the there is a low entropy.
- If the occurances is 50% on each side then the entropy is one.
- Entropy is the measure of the randomness.

Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/salaries.csv')
df
```

Seperating the variables

```python
features = df.drop('salary_more_then_100k', axis='columns')
target = df['salary_more_then_100k']
```

```python
features
```

```python
target
```

Crearing new columns with integers instread of strigns

```python
Le_company = LabelEncoder()
Le_job = LabelEncoder()
Le_degree = LabelEncoder()
```

```python
features['company_n'] = Le_company.fit_transform(features['company'])
features['job_n'] = Le_job.fit_transform(features['job'])
features['degree_n'] = Le_degree.fit_transform(features['degree'])
features.head()
```

Droping previous columns with values in the form of strings

```python
features_n = features.drop(['company', 'job', 'degree'], axis='columns')
features_n.head()
```

Training the model

```python
model_dt = tree.DecisionTreeClassifier()
```

```python
model_dt.fit(features_n, target)
```

```python
model_dt.score(features_n, target)
```

The score is 1 because we use the same dataset for training and testing purpose.

Make Predictions

```python
model_dt.predict([[2, 0, 1]])
```
